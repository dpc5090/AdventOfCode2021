```python
def get_neighbor_pos(x,y):
    neighbors = [[-1,0],[1,0],[0,-1],[0,1]]
    ret = []
    for n in neighbors:
        new_pos = [x+n[0],y+n[1]]
        if 0<=new_pos[0]<len(data):
            if 0<=new_pos[1]<len(data):
                ret.append(new_pos)
    return ret
```
