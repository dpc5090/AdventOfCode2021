```python
f = open("day6.txt",'r')
data = f.read()
data = data.split(",")
```

```python
data = [int(d) for d in data]
data
```

```python
count = Counter(data)
count
```

```python
d= count.copy()
new_d = {}
for i in range(80):
    new_d = {}
    for key in d:
        if key-1 <0:
            new_d[8]  = d[0]
            val = new_d.get(6,0)
            new_d[6] = val+d[0]
        else:
            val = new_d.get(key-1,0)
            new_d[key-1] = val+d[key]
    d = new_d.copy()
sum(d.values())      
```
