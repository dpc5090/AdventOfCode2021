```python
f = open("day2.txt",'r')
data = f.readlines()
```

```python
data = [item.strip().split(" ") for item in data]
```

```python
data
```

```python
depth, horizontal = 0,0
for direction, dist in data:
    if direction == 'forward':
        horizontal+=int(dist)
    elif direction == 'down':
        depth += int(dist)
    elif direction == "up":
        depth -= int(dist)
    else:
        print(direction)
print(depth,horizontal)
depth*horizontal
```

```python
depth, horizontal,aim = 0,0,0

for direction, dist in data:
    if direction == 'forward':
        horizontal+=int(dist)
        depth+=aim*int(dist)
    elif direction == 'down':
        aim += int(dist)
    elif direction == "up":
        aim -= int(dist)
    else:
        print(direction)
print(depth,horizontal)
depth*horizontal
```
