```python
def explode(str_in, last_open, last_close):
    ret_str = ""
    left_number,right_number = str_in[last_open+1:last_close].split(",")
    left_number = int(left_number)
    right_number = int(right_number)
    
    
    #find left digit
    next_left_number = ""
    left_fill = ""
    for x, c in enumerate(str_in[:last_open-1][::-1]):
        if c.isdigit():
            next_left_number = c+next_left_number
        else:
            
            if next_left_number != "":
                break
            left_fill+=c
    #find right digit
    next_right_number = ""
    right_fill = ""
    for y, c in enumerate(str_in[last_close+1:]):
        if c.isdigit():
            next_right_number += c
        else:
            if next_right_number != "":
                break
            right_fill+=c
    ret_str = str_in
    if next_right_number != "":
        ret_str = f"{ret_str[:last_close+y-len(next_right_number)+1]}{int(next_right_number)+right_number}{ret_str[last_close+y+1:]}"
        pass
    
    ret_str = f"{ret_str[:last_open]}{0}{ret_str[last_close+1:]}"
    if next_left_number!= "":

        ret_str = f"{ret_str[:last_open-1-x]}{int(next_left_number)+left_number}{ret_str[last_open-1-x+len(next_left_number):]}"

        pass
    
    
    return ret_str
```

```python
def split(str_in, number):
    i_num = int(number)
    str_in = str_in.replace(number,f"[{int(i_num/2)},{int((i_num+1)/2)}]",1)
    return str_in

split(str_in,"15")
```

```python
def solve(str_in):
    change = True
    while change:
        change = False
        exploding = True
        while exploding:
            open_count = 0
            running_number = ""
            last_open = -1
            last_close = -1
            exploding = False
            for i, c in enumerate(str_in):
                if c.isdigit():
                    running_number+=c
                else:
                    if running_number !="":
                        if len(running_number)>2:
                            print("split")
                        running_number = ""
                    if c == "[":
                        open_count+=1
                        last_open = i
                    if c == "]":
                        last_close = i
                        if open_count > 4:
                            str_in = explode(str_in, last_open, last_close)
                            exploding = True
                            change = True
                            break
                        open_count-=1

        running_number = ""
        for i, c in enumerate(str_in):
            if c.isdigit():
                running_number+=c
            else:
                if running_number !="":
                    if len(running_number)>1:
    
                        str_in = split(str_in, running_number)
                        change = True
                        break
                    running_number = ""
#                     running_number = ""
    return(str_in)
```

```python
val =""
for line in data:
    if val == "":
        val = line
    else:
        val = f"[{val},{line}]"
        print(val)
        val = solve(val)
        print("Solved",val)
```

```python
def get_mag(str_in):
    last_open = 0
    last_close = 0
    keep_going = True
    while keep_going:
        keep_going = False
        for i, c in enumerate(str_in):

            if c == "[":
                last_open = i
            if c == "]":
                last_close = i
                left_number,right_number = str_in[last_open+1:last_close].split(",")
                left_number = int(left_number)
                right_number = int(right_number)
                new_val = left_number*3+right_number*2
                str_in = str_in.replace(str_in[last_open:last_close+1], str(new_val))
                keep_going = True
                break
    return(int(str_in))
get_mag(val)
```

```python
best_val = 0
for row in data:
    for row2 in data:
        if row != row2:
            val = solve(f"[{row},{row2}]")
            best_val = max(get_mag(val), best_val)
best_val
```
