```python
def AlreadyGood(cur_pos, position):
    val = position[cur_pos]
    dest_col = final_map[val]
#     print(cur_pos[1], dest_col)
    if cur_pos[1] != dest_col:
        return False
    
    for i in range(cur_pos[0], 5):
        if position[i,dest_col] != val:
            return False
    return True
```

```python
def getMoveCost(cur_pos, dest_pos, position):
    val = position[cur_pos]
    cost_per_step = cost_dict[val]
    x_dist = abs(cur_pos[1]-dest_pos[1])
    if cur_pos[0] == 0 or dest_pos[0]  == 0:
        y_dist = abs(cur_pos[0]-dest_pos[0])
    elif cur_pos[1]  != dest_pos[1]:
        y_dist = cur_pos[0]+dest_pos[0]
        
    
    
    return (x_dist+y_dist)*cost_per_step
# show(position)
# getMoveCost((1,8),(2,6), position)
```

```python
def getScore(position):
    cost=0
    for pos in position:
        if position[pos] != ".":
            val = position[pos]
            dest_col = final_map[val]
            cost_per = cost_dict[val]
            cost += cost_per*abs(dest_col-pos[1])
    return cost
```

```python
def allDone(position):
    for val in "ABCD":
        dest_col = final_map[val]
#         print(val, dest_col)
        if position[(1,dest_col)] != val or position[(2,dest_col)] != val:
            return False
    return True
```
