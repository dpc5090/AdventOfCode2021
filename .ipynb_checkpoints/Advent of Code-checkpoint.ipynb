```python
def destMoves(cur_pos, position ):
    
    val = position[cur_pos]
    col = final_map[val]
    if cur_pos[1] == col:
        if cur_pos[0] == 2:
            return []
    if cur_pos[1] == col:
        if cur_pos[0] == 1 and position[(2,col)] == val:
            return []
    
    if position[(2,col)] =="." and position[(1,col)] == ".":
        return[ (2,col)]
    if position[(2,col)] ==val and position[(1,col)] == ".":
        return [(1,col)]
    return []
    
```

```python
def AlreadyGood(cur_pos, position):
    val = position[cur_pos]
    dest_col = final_map[val]
    if cur_pos == (2,dest_col):
        return True
    if cur_pos == (1,dest_col) and position[(2, dest_col)] == val:
        return True
    return False
```

```python
def getMoveCost(cur_pos, dest_pos, position):
    val = position[cur_pos]
    cost_per_step = cost_dict[val]
    x_dist = abs(cur_pos[1]-dest_pos[1])
    if cur_pos[0] == 0 or dest_pos[0]  == 0:
        y_dist = abs(cur_pos[0]-dest_pos[0])
    elif cur_pos[1]  != dest_pos[1]:
        y_dist = cur_pos[0]+dest_pos[0]
        
    
    
    return (x_dist+y_dist)*cost_per_step
# show(position)
# getMoveCost((1,8),(2,6), position)
```

```python
def getPossibleMoves(cur_pos, position):
    val = position[cur_pos]
    pos_moves = []
    if AlreadyGood(cur_pos, position):
#         print("all good")
        return []
    if cur_pos[0] == 2:
        if position[(1,cur_pos[1])] != ".":
            return []
        thing = cur_pos[1]
        for j in [ x for x in range(thing)][::-1]:
            if position[(0,j)] == ".":
                pos_moves.append((0,j))
            else:
                break
        for j in [x for x in range(thing+1,11)]:
            if position[(0,j)] == ".":
                pos_moves.append((0,j))
            else:
                break
    elif cur_pos[0] == 1:
        thing = cur_pos[1]
        for j in [ x for x in range(thing)][::-1]:
            if position[(0,j)] == ".":
                pos_moves.append((0,j))
            else:
                break
        for j in [x for x in range(thing+1,11)]:
            if position[(0,j)] == ".":
                pos_moves.append((0,j))
            else:
                break
    else:
        thing = cur_pos[1]
        for j in [ x for x in range(thing)][::-1]:
            
            if position[(0,j)] == ".":
                pos_moves.append((0,j))
            else:
                break
        for j in [x for x in range(thing+1,11)]:
            if position[(0,j)] == ".":
                pos_moves.append((0,j))
            else:
                break
    if (0, final_map[val]) in pos_moves:
        pos_moves+= destMoves(cur_pos, position)
    
    if cur_pos[0] == 0:
        temp = pos_moves.copy()
        for pos in pos_moves:
            if pos[0] == 0:
                temp.remove(pos)
        pos_moves = temp
            
    for pos in pos_moves:
        if pos in [(0,2),(0,4),(0,6),(0,8)]:
            pos_moves.remove(pos)
        
    return pos_moves
# pos_moves = getPossibleMoves((0,5),position)
```
