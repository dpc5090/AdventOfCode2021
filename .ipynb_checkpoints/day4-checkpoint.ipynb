```python
from collections import Counter, deque
```

```python
import numpy as np
import pandas as pd
import heapq
```

```python
f = open("day3.txt",'r')
data = f.readlines()
```

```python
data = [item.strip() for item in data]
```

```python
data
```

```python
d  = {}
for item in data:
    for i, letter in enumerate(item):
        if i in d:
            d[i] += letter
        else: 
            d[i] = [letter]
gamma, epsilon = "",""

for key in d:
    c = Counter(d[key])
    print(key,c)
    gamma += c.most_common()[0][0]
    epsilon += c.most_common()[1][0]
print(gamma, epsilon)
decoded_gamma,decoded_epsilon = int(gamma,2), int(epsilon,2)
decoded_epsilon*decoded_gamma
```

```python
data_df = pd.DataFrame([[int(i)for i in x] for x in data])
```

```python
data_df
```

```python
gamma_df = data_df.copy()
epsilon_df = data_df.copy()
gamma,epsilon = "",""
for i in range(data_df.shape[1]):
    ones = 0
    zeroes = 0
    print(i)
    ones = (gamma_df.iloc[:,i] == 1).sum()
    zeroes = (gamma_df.iloc[:,i] == 0).sum()
    if ones>=zeroes:
        gamma+="1"
        gamma_criterion = 1
    else:
        gamma+="0"
        gamma_criterion = 0
    gamma_df =gamma_df[ gamma_df.iloc[:,i] == gamma_criterion]

    ones = (epsilon_df.iloc[:,i] == 1).sum()
    zeroes = (epsilon_df.iloc[:,i] == 0).sum()
    if ones<zeroes:
        epsilon+="1"
        epsilon_criterion = 1
    else:
        epsilon+="0"
        epsilon_criterion = 0
    if min(ones,zeroes)>0:
        epsilon_df =epsilon_df[ epsilon_df.iloc[:,i] == epsilon_criterion]

    
    print(gamma, epsilon,len(gamma_df),len(epsilon_df))
    print("gamma")
    print(gamma_df.head())
    print("eps")
    print(epsilon_df.head())
```

```python
gamma = "".join(gamma_df.values[0].astype(str))
epsilon = "".join(epsilon_df.values[0].astype(str))
print(gamma, epsilon)
int(gamma,2)* int(epsilon,2)
```

```python
d  = {}
for item in data:
    for i, letter in enumerate(item):
        if i in d:
            d[i] += letter
        else: 
            d[i] = [letter]
gamma, epsilon = "",""

for key in d:
    c = Counter(d[key])
    print(key,c)
    gamma += c.most_common()[0][0]
    epsilon += c.most_common()[1][0]
print(gamma, epsilon)
decoded_gamma,decoded_epsilon = int(gamma,2), int(epsilon,2)
decoded_epsilon*decoded_gamma
```

```python
depth, horizontal = 0,0
for direction, dist in data:
    if direction == 'forward':
        horizontal+=int(dist)
    elif direction == 'down':
        depth += int(dist)
    elif direction == "up":
        depth -= int(dist)
    else:
        print(direction)
print(depth,horizontal)
depth*horizontal
```

```python
depth, horizontal,aim = 0,0,0

for direction, dist in data:
    if direction == 'forward':
        horizontal+=int(dist)
        depth+=aim*int(dist)
    elif direction == 'down':
        aim += int(dist)
    elif direction == "up":
        aim -= int(dist)
    else:
        print(direction)
print(depth,horizontal)
depth*horizontal
```
