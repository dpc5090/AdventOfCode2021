```python
data = [np.array([[int(item) for item in row.split(",")] for row in scanner.split("\n")[1:] ])for scanner in data]
```

```python
import math
```

```python
dist_dict = {}
for scan_id, scanner in enumerate(data):
    dist_dict[scan_id] = {}
    for i in range(len(scanner)):
        for j in range(len(scanner)):
            if i!=j:
                if i<j:
                    dist_dict[scan_id][f"{i}_{j}"] = math.dist(scanner[i,:], scanner[j,:])
                    
        pass
```

```python
stop = False
matched = []
for key, d1 in dist_dict.items():
    for key2, d2 in dist_dict.items():
        if key == key2:
            continue
        pairs = []
        for sub_key, dist1 in d1.items():
            
            if dist1 in d2.values():
                if Counter(d2.values())[dist1]>1:
                    continue
                count = 0
                for sub_key2, dist2 in d2.items():
                    if dist2 == dist1:
                        pairs.append([sub_key, sub_key2])
                        
      
        if len(pairs)>0:
            left = [item[0] for row in pairs for item in row[0].split("_") ]
            lc = Counter(left)
            for l_key in lc:
                sub_rows = []
                for row in pairs:
                    if l_key in row[0].split("_"):
                        sub_rows.append(row[1])
                print(l_key, sub_rows)
                if len(sub_rows)>1:
                    temp_c = Counter([item  for line in sub_rows for item in line.split("_")])
                    for r_key, val in temp_c.items():

                        if val == len(sub_rows):
                            matched.append([key,l_key, key2, r_key])
len(matched)
```

```python
matched_dict = {}
for k1, v1, k2,v2 in matched:
    key = f"{k1}_{v1}"
    key2 = f"{k2}_{v2}"
    if key in matched_dict:
        matched_dict[key]+=[key2]
    else:
        matched_dict[key] = [key2]
    if key2 in matched_dict:
        matched_dict[key2]+=[key]
    else:
        matched_dict[key2] = [key]
matched_dict
```

```python
rows = [[1,0,0],[0,1,0],[0,0,1]]
ms = []
for r1 in rows:
    for r2 in rows:
        for r3 in rows:
            if r1 == r2:
                continue
            if r1 == r3:
                continue
            if r2 == r3:
                continue
            matrix = np.array([r1,r2,r3])
            base = np.array(matrix)
      
            for m1 in [-1,1]:
                for m2 in [-1,1]:
                    for m3 in [-1,1]:
                        matrix[0,:] = base[0,:]*m1
                        matrix[1,:] = base[1,:]*m2
                        matrix[2,:] = base[2,:]*m3
                        ms.append(matrix.copy())
                        
            
            
```

```python
transformed_data = data.copy()
```

```python
fixed = ['0']
searched = []
centers  = {}
centers['0'] = np.array([0,0,0])
while len(fixed)!= len(data):
    k1 = None
    for node in fixed:
        if node not in searched:
            k1 = node
    if k1 == None:
        print("ERROR")
    print("k1",k1)
    searched.append(k1)
    selected_vals = []
    for key, val in matched_dict.items():
        val = list(set(val))
        if f"{k1}_" == key[:len(k1)+1]:
            selected_vals+=[row.split("_")[0] for row in val]
    c = Counter(selected_vals)
    print(selected_vals)
    s1 = transformed_data[int(k1)]
#     print("center",centers[k1])
    for k2 in c:
        if k2 in fixed:
            continue
        print("matching beacons", k1, k2)
        s2 = data[int(k2)]
        pairs = []
        for m,v in matched_dict.items():
            if f"{k1}_" == m[:len(k1)+1]:
                for item in v:

                    if item[:len(k2)+1] == k2+"_":
                        x,y = m.split("_")[1],item.split("_")[1]
                        if [int(x),int(y)] not in pairs:
                            pairs.append([int(x),int(y)])
                    if len(pairs) ==2:
                        break
            if len(pairs) == 2:

                print(pairs)
                for i,m in enumerate(ms):
                    temp = np.matmul(s2, m)
                    if np.all(s1[pairs[0][0]]-s1[pairs[1][0]] == temp[pairs[0][1]]-temp[pairs[1][1]]):
                        break
                
                centers[k2] = s1[pairs[0][0]]-temp[pairs[0][1]]

                t2 = temp+ s1[pairs[0][0]]-temp[pairs[0][1]]
#                 print("center",k2, centers[k2])
                print("k2",k2)
                print("counts",c[k2])
                thing = 0
                for i,x in enumerate(s1):
                    for j, y in enumerate( t2):
                        if np.all(x == y):
                            print("matched",i,j)
                            thing+=1
                            pass
                if thing != c[k2]:
                    print("SOMETHING WEIRD HAPPENED",c[k2],thing)
                transformed_data[int(k2)] = t2.copy()
                fixed.append(k2)
                break
    print("DONE",k1,len(fixed), fixed)
    print()
    
```
