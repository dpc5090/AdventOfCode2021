```python
connections = {}
for row in data:
    x,y = row.split("-")
    if x in connections:
        connections[x]+=[y]
    else:
        connections[x] = [y]
    if y in connections:
        connections[y] += [x]
    else:
        connections[y] = [x]

connections
```

```python
def getNextNodes(node, path):
    possibleNodes  = connections[node]
    nextNodes = []
    for node in possibleNodes:
        if node in path:
            if node.islower():
                continue
        nextNodes.append(node)
    return nextNodes
```

```python
paths = deque([['start',[]]])
```

```python
final_paths = []
while len(paths)>0:
    node, path = paths.pop()
    if node == "end":
        path.append(node)
        final_paths.append(path)
        continue
    nextNodes = getNextNodes(node,path)
    path.append(node)
    for n in nextNodes:
        paths.append([n,path.copy(),])
len(final_paths)
```

```python
def getNextNodesV2(node, path):
    cur_node = node
    possibleNodes  = connections[node]
    nextNodes = []

    for node in possibleNodes:
        if node not in path:
            nextNodes.append(node)
            continue
        else:
            if node.islower() == False:
                nextNodes.append(node)
                continue
            else:
                if node == "start":
                    continue
                c = Counter(path)
                dont = False
                for key in c:
                    if key.islower():

                        if c[key]>1:
                            dont = True
                            break
                if dont == True:
                    continue
                else:
                    print(path, node,cur_node)
                    nextNodes.append(node)
                    continue

    return nextNodes
```

```python
paths = deque([['start',[]]])
final_paths = []
while len(paths)>0:
    node, path = paths.pop()
    if node == "end":
        path.append(node)
        final_paths.append(path.copy())
        continue
    nextNodes = getNextNodesV2(node,path)
    path.append(node)
    for n in nextNodes:
        paths.append([n,path.copy()])
len(final_paths)
```
